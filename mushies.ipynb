# Run a found ND

```python
import numpy as np
import config

from sklearn.linear_model import LogisticRegression

def get_or_train_split_model(left, right):
    key = (left, right)
    if key in config.model_cache:
        return config.model_cache[key]

    # use global training data
    X_tr = config.X
    y_tr = config.y

    # keep only examples belonging to this node
    # (classes in left âˆª right)
    node_classes = tuple(left) + tuple(right)
    mask = np.isin(y_tr, node_classes)
    X_node = X_tr[mask]
    y_node_raw = y_tr[mask]

    # map to 0/1: left -> 0, right -> 1
    is_left = np.isin(y_node_raw, left)
    y_node = np.where(is_left, 0, 1)

    # train logistic regression for this split
    model = LogisticRegression(
        penalty="l2",
        solver="newton-cholesky",
        C=config.C,          # set this somewhere in config, e.g. 0.1
        max_iter=2000
    )
    model.fit(X_node, y_node)

    config.model_cache[key] = model
    return model

def nd_predict_proba(X, T, classes):
    class_to_idx = {c: i for i, c in enumerate(classes)}
    n_samples = X.shape[0]
    n_classes = len(classes)
    proba = np.ones((n_samples, n_classes), dtype=float)

    # multiply probs down the ND
    for left, right in T:
        model = get_or_train_split_model(left, right)
        p_lr = model.predict_proba(X)
        p_left = p_lr[:, 0]
        p_right = p_lr[:, 1]

        li = [class_to_idx[c] for c in left]
        ri = [class_to_idx[c] for c in right]

        proba[:, li] = proba[:, li] * p_left[:, None]
        proba[:, ri] = proba[:, ri] * p_right[:, None]

    # normalise per row
    row_sums = proba.sum(axis=1)
    proba = proba / row_sums[:, None]
    return proba, classes


def nd_predict(X, T, classes):
    proba, classes = nd_predict_proba(X, T, classes)
    idx = proba.argmax(axis=1)
    return np.array(classes)[idx]


def nd_logloss(X, y, T, classes):
    proba, classes = nd_predict_proba(X, T, classes)
    class_to_idx = {c: i for i, c in enumerate(classes)}

    y_arr = np.array(y, dtype=int)
    n = len(y_arr)
    idx = np.array([class_to_idx[int(c)] for c in y_arr], dtype=int)

    eps = 1e-15
    p_true = np.clip(proba[np.arange(n), idx], eps, 1.0)
    return -np.mean(np.log(p_true))
```

```python
# best_trees.index: iterable of trees T
# best_trees.values: corresponding loglik scores

def ensemble_nd_proba(X, best_trees, classes=None):
    scores = best_trees["score"].astype(float)
    trees = list(best_trees["tree"])

    # convert logliks to positive weights (softmax in log space)
    temperature = 100  # try 1.5, 2, 3 etc.

    scores_shift = (scores - scores.max()) / temperature
    weights_unnorm = np.exp(scores_shift)
    weights = weights_unnorm / weights_unnorm.sum()

    proba_ens = None
    for T, w in zip(trees, weights):
        p_T, classes = nd_predict_proba(X, T, classes)
        if proba_ens is None:
            proba_ens = w * p_T
        else:
            proba_ens = proba_ens + w * p_T

    return proba_ens, classes


def ensemble_predict(X, best_trees, classes=None):
    proba, classes = ensemble_nd_proba(X, best_trees, classes)
    idx = proba.argmax(axis=1)
    return np.array(classes)[idx]


def ensemble_logloss(X, y, best_trees, classes=None):
    proba, classes = ensemble_nd_proba(X, best_trees, classes)
    class_to_idx = {c: i for i, c in enumerate(classes)}

    y_arr = np.array(y, dtype=int)
    n = len(y_arr)
    idx = np.array([class_to_idx[int(c)] for c in y_arr], dtype=int)

    eps = 1e-15
    p_true = np.clip(proba[np.arange(n), idx], eps, 1.0)
    return -np.mean(np.log(p_true))
```

```python
import numpy as np

def flatten_traces(traces, burn_in=0, thin=1):
    trees = []
    for chain in traces:
        trees.extend(list(chain[burn_in::thin]))
    return list(dict.fromkeys(trees))


def bma_proba(X, trees, classes=None):
    proba = None
    for T in trees:
        p_T, classes = nd_predict_proba(X, T, classes)
        if proba is None:
            proba = p_T
        else:
            proba = proba + p_T
    proba = proba / float(len(trees))
    return proba, classes


def bma_predict(X, trees, classes=None):
    proba, classes = bma_proba(X, trees, classes)
    idx = proba.argmax(axis=1)
    return np.array(classes)[idx]


def bma_logloss(X, y, trees, classes=None):
    proba, classes = bma_proba(X, trees, classes)
    class_to_idx = {c: i for i, c in enumerate(classes)}

    y_arr = np.array(y, dtype=int)
    idx = np.array([class_to_idx[int(c)] for c in y_arr], dtype=int)

    eps = 1e-15
    p_true = np.clip(proba[np.arange(len(y_arr)), idx], eps, 1.0)
    return -np.mean(np.log(p_true))
```
